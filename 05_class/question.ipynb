### **문제 1: 클래스 속성, 인스턴스 속성 테스트**

카피바라는 이직 준비를 위해 자기소개서를 작성해야 합니다. `Capybara`이라는 클래스를 만들어 카피바라의 기본 정보를 설정하세요.

1. 클래스 속성 `species`를 설정하여 "카피바라"로 초기화하세요.
2. 생성자를 통해 인스턴스 속성 `name`과 `current_job`을 설정하세요.
3. 메서드 `introduce`를 만들어 이름과 현재 직업을 소개하는 문자열을 반환하세요.

```python
class Capybara:
    species = '카피바라'
    def __init__(self, name, current_job):
        self.name = name
        self.current_job = current_job

    def introduce(self):
        return '제 이름은 ' + self.name + '입니다. 제 species는 ' + self.species + '입니다. 제 저의 현재 직업은 ' + self.current_job + '입니다.'
capybara = Capybara('카피바라', '학생')
capybara2 = Capybara('카카','카카')
print(capybara.introduce())
print(capybara2.introduce())
```

### **문제 2: 클래스 속성과 인스턴스 속성의 차이**

카피바라는 이직 준비를 하며 다양한 직업 스킬을 학습 중입니다. 클래스 속성과 인스턴스 속성의 차이를 확인하세요.

1. `Character` 클래스를 생성하세요.
2. 클래스 속성 `skills`를 정의하여 빈 리스트로 초기화하세요.
3. 생성자에서 인스턴스 속성 `skills`를 별도로 초기화하세요.
4. `add_skill` 메서드에서 클래스 속성이 아닌 인스턴스 속성에 스킬을 추가하도록 수정하세요.


```python
class Character:
    skills = []
    
    def _init__(self):
        self.skills = []

    def add_skill(self, skill):
        self.skills.append(skill) #자기 자신의 skills 속성에 접근
        
character1 = Character()
character2 = Character()

character1.add_skill('금강불괴')
character2.add_skill('날라차기')

print(character1.skills) # ['금강불괴']
print(character2.skills) # ['날라차기']
```

### **문제 4: 네임스페이스 테스트 - 전역과 로컬**

카피바라는 이직 준비 도구를 사용하며, 전역 변수와 로컬 변수를 관리합니다. `namespace_test` 함수를 작성해 네임스페이스를 테스트하세요.

1. 전역 변수 `tool`을 "연필"로 설정하세요.
2.  함수 `namespace_test` 내부에 로컬 변수 `local_tool`을 "키보드"로 설정하고, `global` 키워드를 사용해 전역 변수를 "펜"으로 수정하세요.
3. 함수 내부에서 로컬 변수와 전역 변수의 값을 출력하세요.

```python
tool = '연필'

def namespace_test():
    local_tool = '키보드'
    global tool
    tool = '펜'
    print(local_tool)     
    print(tool)

namespace_test()
print(tool)
```

### **문제 5: 네임스페이스 테스트 - 클래스와 인스턴스**

카피바라는 이직 준비 중 클래스를 통해 도구를 관리합니다. `Toolbox` 클래스를 만들어 네임스페이스를 테스트하세요.

1. 클래스 속성 `tools`를 ["노트북", "책상"]으로 초기화하세요.
2. 생성자를 통해 인스턴스 속성 `tools`를 설정하고, 객체마다 다른 도구를 가질 수 있도록 설정하세요.
3. `add_tool` 메서드를 만들어 인스턴스 속성에 새 도구를 추가하세요.

```python
class Toolbox:
    tools = ["노트북","책상"]

    def __init__(self):
        self.tools = []

    def add_tool(self, tool):
        self.tools.append(tool)

toolbox1 = Toolbox()
toolbox2 = Toolbox()

toolbox1.add_tool("연필")
toolbox2.add_tool("지우개")
toolbox1.add_tool("펜")
toolbox2.add_tool("자")

print(toolbox1.tools) # ['연필', '펜']
print(toolbox2.tools) # ['지우개', '자']

print(Toolbox.tools) # ['노트북', '책상']
```

### **문제 6: 네임스페이스 테스트 - nonlocal**

카피바라는 이직 준비 중 추가적인 정보가 필요합니다. 중첩 함수와 `nonlocal` 키워드를 사용해 바깥 함수의 변수를 수정하세요.

1. 함수 `nested_function`을 작성하세요.

1. 바깥 함수의 변수 `status`를 "작성 중"으로 초기화하세요.
2. 중첩 함수에서 `status`를 "완료"로 수정하세요.

```python
def nested_function():
    status = "작성 중"

    def inner_function():
        nonlocal status
        status = "완료"
    print(status)
    inner_function()
    print(status)
nested_function()
```

### 문제 1: **기본 상속을 활용한 클래스 생성**

---

카피바라는 이직을 준비하며 새로운 직업군을 탐색하고 있습니다. 카피바라의 기본적인 정보를 담는 `Person` 클래스를 만들고, 이를 상속받는 `Candidate` 클래스를 생성하세요.

1. `Person` 클래스:
    - 클래스 속성 `national`은 "Korea"로 설정합니다.
    - 메서드 `introduce`는 `"안녕하세요, 저는 한국에서 왔습니다."`를 반환합니다.
2. `Candidate` 클래스:
    - `Person` 클래스를 상속받습니다.
    - 생성자에서 이름(`name`)과 관심 분야(`field`)를 매개변수로 받아 저장합니다.
    - 메서드 `introduce`를 오버라이드하여 `"안녕하세요, 저는 [name]입니다. [field] 분야에 관심이 있습니다."`를 반환합니다.

```python
class Person:
    national = 'korea'

    def introduce(self):
        return("안녕하세요, 저는 한국에서 왔습니다.")

class Candidate(Person):
    def __init__(self, name, field):

    self.name = name
    self.field = field

    def introduce(self):
        return('안녕하세요, 저는 'self.name'입니다. '+self.field+'분야에 관심이 있습니다.')

candidate = Candidate()
print(candidate.introduce())
```
