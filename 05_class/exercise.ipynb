```python
g_variable = "global variable"

def modify_global():
    global g_variable
    g_variable = "global modified"

print(g_variable)    # global variable
modify_global()
print(g_variable)    # global modified
```

```python
def outer_function():
    variable = "outer variable"

    def inner_function():
        nonlocal variable
        variable = "modified"

    print(variable)
    inner_function()
    print(variable)

outer_function()
```

```python
class Character:
    def __init__(self, nickname, type):
        self.nickname = nickname
        self.__type = type
    
    def get_type(self):
        return self.__type

    def set_type(self, type):
        character_types= ['전사', '법사', '힐러']
        if type in character_types:
            self.__type = type
        else:
                raise ValueError("잘못된 캐릭터 타입입니다.")

my_character = Character("카피바라", "전사")

print(my_character.nickname)
#print(my_character.__type)

print(my_character.get_type())

my_character.set_type("법사")
print(my_character.get_type())

# my_character.set_type("흑법사")


```

```python
class Person:    
    national = 'korea'
    language = 'korean'
    
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greeting(self):
        return '안녕하세요'
    
    def information(self):
        return "I'm from " + self.national + " and I use " + self.language
     
    def favorite(self, color):
        return "I love " + color

squirrel = Person('다람쥐', 900)
capybara = Person('카피바라', 20)

print(squirrel)    # <__main__.Person object at 0x000002E16EE94560>
print(capybara)     # <__main__.Person object at 0x000002E16EE94BF0>

print(squirrel.national)            # korea
print(capybara.national)            # korea
print(squirrel.greeting())          # 안녕하세요
print(capybara.greeting())          # 안녕하세요

print(squirrel.information())        # I'm from korea and I use korean. My name is 다람쥐. I'm 900
print(capybara.information())        # I'm from korea and I use korean. My name is 카피바라. I'm 20

print(capybara.favorite('black'))    # I love black
print(squirrel.favorite('blue'))     # I love blue
```

```python
class Character:    
    skills = []
    
    def __init__(self, nickname, type):
        self.nickname = nickname
        self.type = type
        
    def add_skill(self, skill):
        self.skills.append(skill) #자기 자신의 skills 속성에 접근
    
    def show_skill(self):
        return self.type + ' ' + self.nickname + '님의 보유 스킬 ' + str(self.skills)

my_character = Character('산골짜기다람쥐', '전사')
your_character = Character('도시카피바라', '법사')
my_character.add_skill('곤봉 휘두르기')
your_character.add_skill('마법진 그려서 날려버리기')

print(my_character.show_skill())    # 전사 산골짜기다람쥐님의 보유 스킬 ['곤봉 휘두르기', '마법진 그려서 날려버리기']
```

```python
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, grade):
        super().__init__(name)
        self.grade = grade

student = Student("카피바라", "2학년")
print(student.name)    # 카피바라
print(student.grade)   # 2학년
```

```python
class Person:
    national = 'korea'

    def greeting(self):
        return 'this is python'

class Learner:
    def greeting(self):
        return 'hello i am learner'

    def learn(self):
        return 'i am learning python'

class Student(Person, Learner):
    pass

student1 = Student()
print(student1.greeting())    # this is python
print(student1.learn())       # i am learning python
```

```python
class Person:
    national = 'korea'

    def greeting(self):
        return 'this is python'

class Learner(Person):
    def __init__(self,subject):
        self.subject = subject

    def learn(self):
        return "i am learning " + self.subject

class Student(Learner):
    def __init__(self, name, subject):
        super().__init__(subject)
        self.name = name

    def greeting(self):
        return "hello i am " + self.name 
student = Student('카피바라', 'python')
print(student.greeting())    # this is python
print(student.learn())       # i am learning python
```

```python
class Animal:
    def do(self):
        print('👉👈')
class Dog(Animal):
    def do(self):
        print('🐶🐶🐶')
class Duck(Animal):
    def do(self):
        print('🦆🦆🦆')
class Fish(Animal):
    pass    # do 호출시 부모 클래스 Animal#do가 호출된다.

for each in (Dog(), Duck(), Fish()):
    each.do()
```

```python
# 상속 관계가 아니더라도 동일한 이름의 함수가 있는 경우 다형성처럼 사용이 가능
class Alarm:
    def do(self):
        print('⏰⏰⏰')

for each in (Dog(), Duck(), Fish(), Alarm()):
    each.do()
```
