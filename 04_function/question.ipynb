## **문제 3: 업무 시간 계산하기**

**"카피바라는 오늘 근무한 시간을 계산하고 싶다."**

1. `work_hours(start, end)`라는 함수를 만들어 매개변수 `start`(출근 시간)와 `end`(퇴근 시간)를 입력받아 근무 시간을 계산하세요.
    - 근무 시간 = `end - start`
2. 반환된 근무 시간을 출력하세요.
    - 출력 예시: **"오늘 근무한 시간은 [근무 시간]시간입니다."**

```python
def work_hours(start, end):
    return end - start

start = int(input("출근시간:"))
end = int(input("퇴근시간:"))

print(f"오늘 일한 시간은 {work_hours(start, end)}시간입니다.")
```

## **문제 4: 점심 메뉴 추천**

**"점심시간, 카피바라는 팀원들에게 점심 메뉴를 추천해준다."**

1. `recommend_menu(*menus)`라는 함수를 만들어 여러 개의 점심 메뉴를 매개변수로 전달받으세요.
2. 랜덤하게 하나의 메뉴를 선택하고 반환하세요.
    - 출력 예시: **"오늘의 점심 메뉴는 [선택된 메뉴]입니다!"**
    - 힌트: `random.choice()`를 사용하면 랜덤 선택이 가능합니다.

```python
import random

def random_menu(*menu):
    menu = random.choice(menu)
    return menu

menu = random_menu("김밥", "라면", "떡볶이", "순대")
print(f'오늘의 메뉴는 {menu}입니다.')
```

## **문제 5: 오후 커피 마시기**

**"오후에 졸린 카피바라는 커피를 마시려고 한다."**

1. `coffee_time()`이라는 함수를 만들어 다음과 같은 문자열을 반환하세요.
    - **"오후 3시입니다. 커피 한 잔 하세요!"**
2. 반환된 문자열을 출력하세요.

```python
def coffee_time(time):
    if time < 12:
        print(f"아침 {time}시입니다. 커피 한 잔 하세요!")
    elif 12 <= time < 18:
        print(f"오후 {time}시입니다. 커피 한 잔 하세요!")
    else:
        print(f"저녁 {time}시입니다. 커피 한 잔 하세요!")
coffee_time(15)
```

## **문제 6: 업무 성과 분석하기**

**"카피바라는 오늘의 업무 성과를 보고해야 한다."**

1. `work_report(tasks_done)`이라는 함수를 만들어 오늘 처리한 업무 수를 매개변수로 받으세요.
2. 함수는 다음 두 값을 반환합니다.
    - `tasks_done`: 처리한 업무 수
    - `status`: 업무 성과 평가 (10개 이상: "우수", 5~9개: "보통", 4개 이하: "노력 필요")
3. 반환된 값을 사용해 다음 메시지를 출력하세요.
    - **"오늘 처리한 업무는 [업무 수]개입니다. 성과 평가: [평가]"**
        
        ---

```python
def work_report(tasks_done):

    status =""
    if tasks_done >= 10 :
        status = "우수"
    elif 5 <= tasks_done < 10:
        status = "보통"
    else:
        status = "노력 필요"
    return tasks_done, status
        
tasks_done, status = work_report(7)

print(f"오늘 완료한 업무는 {tasks_done}개입니다. 업무 상태: {status}")

```

## **문제 7: 월급 계산 및 세금 공제**

**"카피바라는 첫 월급을 계산하고 세금을 공제해야 한다."**

1. `calculate_salary(base_salary, bonus)`라는 함수를 만들어 기본 월급과 보너스를 매개변수로 받으세요.
2. 총 월급을 계산하고, 세금은 총 월급의 10%로 공제하세요.
3. 함수는 다음 두 값을 반환합니다.
    - `net_salary`: 실수령액 (총 월급 - 세금)２
    - `tax`: 세금 공제액
4. 반환된 값을 사용해 다음 메시지를 출력하세요.
    - **"세전 월급: [총 월급]원, 세금: [세금]원, 실수령액: [실수령액]원"**

```python
def calculate_salary(base_salary, bonus):

    total_salary = base_salary + bonus
    tax = total_salary * 0.1
    net_salary = total_salary - tax

    return total_salary, tax, net_salary

result = calculate_salary(3000000000, 500000)
    
print(f"세전 월급: {int(result[0])}원, 세금: {int(result[1])}원, 실수령액: {int(result[2])}원")
```
