```python
def calc(a, b):
    return a + b, a - b

print(calc(2, 3))  
print(type(calc(2, 3)))
```

```python
plus, minus = calc(2, 3)
print(f'plus 는 {plus}, minus 는 {minus}')
```

```python
def add_many(*args):
    result = 0

    for i in args:
        result = result + i
    
    return result
print(add_many(1, 2, 3, 4, 5))
```

```python
def show(*args):
    print(args)

show(1, 2, 3, "capybara")
```

```python
def show_info(**kwargs):
    print(kwargs)

show_info(name="capybara", color="black", age=5)
```

```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = map(lambda x: x ** 2, numbers )
print(list(squared_numbers))
```

```python
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = filter(lambda x: x % 2 == 0, numbers)
print(list(even_numbers))
```

```python
points = [(1, 2), (3, 1), (5, -1)]
sorted_points = sorted(points, key=lambda x: x[1])
print(sorted_points)
```

```python
students = [
    {"name":"bear", "score":90},
    {"name":"tiger", "score":70},
    {"name":"rabbit", "score":85}
]

# 점수 기준 정렬
print(sorted(students, key=lambda x: x["score"]))
```
