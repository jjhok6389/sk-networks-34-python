```python
def add_many(*args):
    result = 0

    for i in args:
        result = result + i
    
    return result
print(add_many(1, 2, 3, 4, 5))
```

```python
pairs = [('A', 1), ('B', 2), ('C', 3)]

letters, numbers = zip(*pairs)

print(letters)   # ('A', 'B', 'C')
print(numbers)   # (1, 2, 3)
```

```python
print(zip(letters, numbers))        # <zip object at 0x0000019B5ED5FAC0>
print(list(zip(letters, numbers)))  # [('A', 1), ('B', 2), ('C', 3)]
```

```python
## dict 정렬
d = {
    'blockdmask': 400,
     'equal': 300,
     'apple': 200,
     'dish': 100,
     'cook': 500 
     }

print(d)
print('--------------')
print(d.keys())
print('--------------')
print(d.items())

# value 기준
print(sorted(d.items(), key = lambda x : x[1]))

# key 기준
print(sorted(d.items(), key = lambda x: x[0]))
```
